```
matrix_p = np.array([[1, 3],
                     [-1, 2]])
input_vector = np.array([3, 4])

func_calc = np.array([input_vector[0] + 3 * input_vector[1],
                      -input_vector[0] + 2 * input_vector[1]])
matmul_calc = np.matmul(matrix_p,
                        input_vector)

print("Matrix:\n", matrix_p)
print("Input Vector:\n", input_vector)
print("Traditional function output:", func_calc)
print("Vector-matrix multiplication output:", matmul_calc)
```

```
matrix_q = np.array([[4, 1, 2],
                     [-1, 2, 3],
                     [5, 1, -2]])
input_vector = np.array([3, 4, 5])

func_calc = np.array([4 * input_vector[0] + input_vector[1] + 2 * input_vector[2],
                      -input_vector[0] + 2 * input_vector[1] + 3 * input_vector[2],
                      5 * input_vector[0] + input_vector[1] - 2 * input_vector[2]])
matmul_calc = np.matmul(matrix_q,
                        input_vector)

print("Matrix:\n", matrix_q)
print("Input Vector:\n", input_vector)
print("Traditional function output:", func_calc)
print("Vector-matrix multiplication output:", matmul_calc)
```

# Eigenvalues and Eigenvectors

## 5.1 In your own words, give an explanation for the intuition behind eigenvalues and eigenvectors.

An eigenvector is, as demonstrated this morning, somewhat like the earth's axis. It's a line from which all points in a sample/data set are least distant. An eigenvalue is the factor by which an eigenvector is stretched (or compressed), which can be a negative quantity, in which case eigenvector direction is reversed.
