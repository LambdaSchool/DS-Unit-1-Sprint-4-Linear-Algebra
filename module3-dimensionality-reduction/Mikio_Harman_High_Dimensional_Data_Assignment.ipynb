## 4.2 Verify that your transformation matrices are correct by choosing an input matrix and calculating the result both via the traditional functions above and also via vector-matrix multiplication.

```
x1 = 2
x2 = 1
x3 = 9

#Creating my tranformed and input arrays for p
p_transformed = np.array([[1,3],
                          [-1,2]])
p_inputs = np.array([[2],
                     [1]])

#Creating my tranformed and input arrays for q
q_transformed = np.array([[4,1,2],
                          [-1,2,3],
                          [5,1,-2]])
q_inputs = np.array([[2],
                     [1],
                     [9]])

def p(a,b):
  return np.array([[a + 3*b],
                   [2*b - a]])
  
def q(a,b,c):
  return np.array([[4*a + b + 2*c],
                   [2*b - a + 3*c],
                   [5*a - 2*c + b]])  
print('p inputs:')
print(p_inputs)
print()
print('p transformed:')
print(p_transformed)  
print()
print('p function output:')
print(p(x1, x2))
print()
print('matrix multiplication output of p:')
print(np.matmul(p_transformed, p_inputs))
print()
print('q inputs:')
print(q_inputs)
print()
print('q transformed:')
print(q_transformed)  
print()
print('q function output:')
print(q(x1, x2, x3))
print()
print('q matrix multiplication output:')
print(np.matmul(q_transformed, q_inputs))
```

The code below will read in the dataset and perform the one-hot encoding of the categorical variables. Start adding your PCA code at the bottom of the provided code.

```
import os.path
from urllib.request import urlopen
from zipfile import ZipFile
from io import BytesIO
import pandas as pd

# Read National Data 
national_url = 'https://www.huduser.gov/portal/datasets/hads/hads2013n_ASCII.zip'
national_file = 'thads2013n.txt'

if os.path.exists(national_file):
    national = pd.read_csv(national_file)
else: 
    z_national = urlopen(national_url)
    zip_national = ZipFile(BytesIO(z_national.read())).extract(national_file)
    national = pd.read_csv(zip_national)
```
