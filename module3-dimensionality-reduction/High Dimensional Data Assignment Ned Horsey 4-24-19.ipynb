## 1.2 Why are graphs that don't pass the vertical line test not considered "functions?"

A function is defined as a relationship between inputs and outpus so that each input is mapped to a single output. The vertical line test visually represents that a non-function will touch a vertical line twice. The vertical line represents a single input while the 'function' line is the output, so if two outputs are generated it will touch the vertical line twice.

# Eigenvalues and Eigenvectors

## 5.1 In your own words, give an explanation for the intuition behind eigenvalues and eigenvectors.

An eigenvector is a vector that is doesn't change it's orientation during a transformation, but it might be scaled by the transformation so an eigenvalue is the scalor for an eigenvector.

Intuitively, the line around which the earth rotates is an obvious eigenvector, it points 'up' and 'down' but every other vector on the curved surface of the earth has it's oreintation shifted by the rotating surface.

Another eigenvalue could be the trajectory of a bullet, the bullet travels along it's eigenvector but as it slows due to air resistance (a transformation) it's eigenvalue might change so it scales down to hit the earth.

# The Curse of Dimensionality

## 6.1 What are some of the challenges of working with high dimensional spaces?

As the number of dimensions increases, the volume of the dimensional space increases exponentially  resulting in data appearing sparse in this large volume, this results in problems of statistical significance and also might make all objects in the set appear sparse and dissimilar from one another, confounding analysis.
