```
mult_CD = np.matmul(vec_C,vec_D)
mult_AE = np.matmul(vec_A,vec_E)
mult_BA = np.matmul(vec_B,vec_A)

print(f'CD: {mult_CD.shape}\n{mult_CD}\n AE: {mult_AE.shape}\n {mult_AE}\n BA: {mult_BA.shape}\n{mult_BA}')
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
vec_H = np.array([[12,11],[7,10]])
vec_J = np.array([[0,1,2],[7,10,4],[3,2,0]])

det_H = np.linalg.det(vec_H)
det_J = np.linalg.det(vec_J)

print(f'The determinant of H: {np.rint(det_H)}')
print(f'The determinant of J: {np.rint(det_J)}')

      
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
inv_H = np.linalg.inv(vec_H)
inv_J = np.linalg.inv(vec_J)

print(inv_H)
print(inv_J)
```

## 3.4 Find $HH^{-1}$ and then find $J^{-1}J$. Is $HH^{-1} == J^{-1}J$? Why or Why not?

```
H_Hinv = np.matmul(vec_H, inv_H)
print(H_Hinv)

J_Jinv = np.matmul(vec_J, inv_J)
print(J_Jinv)
```

They are not exactly the same, but they return their respective identity matrix, based on the matrix's dimensionality. Vector H is 2-dimensional, while vector J is 3-dimensional. Any matrix multiplied by its inverse results in the identity matrix.
