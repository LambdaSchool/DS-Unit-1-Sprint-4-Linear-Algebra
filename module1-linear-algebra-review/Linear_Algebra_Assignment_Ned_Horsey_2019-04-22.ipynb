## 1.1 Create a two-dimensional vector and plot it on a graph

```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

```python
vector = [1,1]

# Plot it as a Vector arrow
ax = plt.axes()
ax.arrow(0,0, vector[0], vector[1], head_width=.05, head_length=0.05, color ='red')
plt.title("One Vector")
ax.set_xlim(0,5)
ax.set_ylim(0,5)

plt.show()
```

## 1.3 Scale the vectors you created in 1.1 by $5$, $\pi$, and $-e$ and plot all four vectors (original + 3 scaled vectors) on a graph. What do you notice about these vectors? 

```python
vector = [1,1]

# scalors

# Then multiply it by a scalar

green = np.multiply(5, vector)

red = np.multiply(math.pi, vector)

blue = np.multiply(-(math.e), vector)


# Plot the Scaled Vectors
ax = plt.axes()
ax.arrow(0,0, vector[0], vector[1], head_width=.05, head_length=0.05, color ='black')
plt.arrow(0,0, red[0], red[1],head_width=.05, head_length=0.05, color ='red')
plt.arrow(0,0, green[0], green[1],head_width=.05, head_length=0.05, color ='green', alpha=0.5)
plt.arrow(0,0, blue[0], blue[1],head_width=.05, head_length=0.05, color ='blue')

plt.title("Four Vectors")
ax.set_xlim(0,8)
ax.set_ylim(0,8)

plt.show()
```

## 1.5 find $\vec{a} - \vec{b}$ and plot the result on the same graph as $\vec{a}$ and $\vec{b}$. Is there a relationship between vectors $\vec{a} \thinspace, \vec{b} \thinspace \text{and} \thinspace \vec{a-b}$

```python
a = [5,7]
b=[3,4]

ab = np.add(a,b)

# Plot Vectors
ax = plt.axes()
ax.arrow(0,0, a[0], a[1], head_width=.05, head_length=0.05, color ='black')
plt.arrow(0,0, b[0], b[1],head_width=.05, head_length=0.05, color ='red')
plt.arrow(0,0, ab[0], ab[1],head_width=.05, head_length=0.05, color ='orange')


plt.title("Two Vectors")
ax.set_xlim(0,12)
ax.set_ylim(0,12)

plt.show()
```

Looks like a-b is inbetween the two vectors.
