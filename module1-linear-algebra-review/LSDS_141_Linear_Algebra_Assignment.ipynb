## 1.2 Create a three-dimensional vector and plot it on a graph

```python
# Import 3d Graphing Library
from mpl_toolkits.mplot3d import Axes3D

# Define 3d vector
vector_ddd = [6, 7, 2]
vector_origin = [0, 0, 0]
vectotal = vector_origin + vector_ddd

# Convert to numpy array
vector_np = np.array([vectotal])

# Plot in 3d space
X, Y, Z, U, V, W = zip(*vector_np)
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.quiver(X, Y, Z, U, V, W, length=1)

# Set viewable area
ax.set_xlim([0, 10])
ax.set_ylim([0, 10])
ax.set_zlim([0, 10])

# Labels
ax.set_xlabel("Width")
ax.set_ylabel("Height")
ax.set_zlabel("Depth")

plt.show()
```

## 1.5 find $\vec{a} - \vec{b}$ and plot the result on the same graph as $\vec{a}$ and $\vec{b}$. Is there a relationship between vectors $\vec{a} \thinspace, \vec{b} \thinspace \text{and} \thinspace \vec{a-b}$

```python
# This time not both in the same list
a = [5, 7]
b = [3, 4]

# Convert list vectors to numpy array vectors
np_a = np.array(a)
np_b = np.array(b)

# Subtract to create new vector
np_c = np_a - np_b

# Plot all three
plt.axis([0, 10, 0, 10])
plt.arrow(0, 0, np_a[0], np_a[1], head_width=0.2, head_length=0.2, color="r")
plt.arrow(0, 0, np_b[0], np_b[1], head_width=0.2, head_length=0.2, color="b")
plt.arrow(0, 0, np_c[0], np_c[1], head_width=0.2, head_length=0.2, color="purple")
    
# plt.xlim(0, 10)
# plt.ylim(0, 10)

plt.show()
```

```python
np_c
```

As $\vec{a-b}$ is a direct result of the two vectors $\vec{a}$ and $\vec{b}$, there is inherently going to be a relationship. The way I see the relationship is that the result (purple) is the vector that would connect the tips of the two other vectors, if the purple one had its origin at the tip of the blue one.

Another way to say this is that the relationship is that the direction of the purple line is the difference between the direction of the other two, and the same thing with the length.

## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```python
# Define the vectors as numpy arrays
g = np.array([1, 1, 1, 8])
h = np.array([3, 3, 3, 3])
```

```python
# Calculate ||g||
g_norm = np.linalg.norm(g)
g_norm
```

```python
# Calculate ||h||
h_norm = np.linalg.norm(h)
h_norm
```

$||g||$ is longer.
