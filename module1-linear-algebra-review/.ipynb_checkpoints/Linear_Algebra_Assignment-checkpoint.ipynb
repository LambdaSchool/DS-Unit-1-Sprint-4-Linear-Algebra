## 1.3 Scale the vectors you created in 1.1 by $5$, $\pi$, and $-e$ and plot all four vectors (original + 3 scaled vectors) on a graph. What do you notice about these vectors? 

```python
## Import pi, e
from math import pi, e
```

```python
type(b)==np.ndarray
```

```python
# Scale vectors
scale_factors = [5, pi, e]

def scaleVector(vector, scalars):
    '''
    Returns original vector and scalars as nested np.array
    '''
    if type(vector)==np.ndarray:
        vectors = np.array([vector])
        for scalar in scalars:
            vectors = np.append(vectors, [vector*scalar], axis=0)
    else:
        print('Please pass in np.ndarray for vector')
    
    return vectors
    
scaled_vectors = scaleVector(b, scale_factors)
scaled_vectors
```

```python
zip(scaled_vectors) # work in progress, may not be necessary
for vec in scaled_vectors:
    print(vec)
```

```python
# Plotting them all
# Specify figure and axes object(s)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_ylim([0,4])
ax.set_xlim([0,2])
ax.set_zlim([0,7])
ax.set_xticks([0,1,2])
ax.set_yticks([0,1,2,3,4])
ax.set_zticks([0,1,2,3,4,5,6,7])

# Using a for loop to plot them all...quickly.
colors = ['red', 'green', 'orange', 'yellow']
for num, vec in enumerate(scaled_vectors):
    xd = np.array([vec[0],0,0])
    yd = np.array([vec[1],0,0])
    zd = np.array([vec[2],0,0])
    ax.plot3D(xd, yd, zd, color=colors[num])
    
plt.show()
```

**Scaled vectors are all in the same direction some combination of C x unitDirection.**

## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{e} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{f} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```python
from numpy import linalg as LA

e = np.array([1,1,1,8])
f = np.array([3]*4)
print('Length e:', LA.norm(e))
print('Length f:', LA.norm(f))
print(LA.norm(e) > LA.norm(f))
```

# Part 2 - Matrices

## 2.1 What are the dimensions of the following matrices? Which of the following can be multiplied together? See if you can find all of the different legal combinations.
\begin{align}
A = \begin{bmatrix}
1 & 2 \\
3 & 4 \\
5 & 6
\end{bmatrix}
\qquad
B = \begin{bmatrix}
2 & 4 & 6 \\
\end{bmatrix}
\qquad
C = \begin{bmatrix}
9 & 6 & 3 \\
4 & 7 & 11
\end{bmatrix}
\qquad
D = \begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}
\qquad
E = \begin{bmatrix}
1 & 3 \\
5 & 7
\end{bmatrix}
\end{align}

```python
A = np.mat([[1,2],[3,4],[5,6]])
B = np.mat([2,4,6])
C = np.mat([[9,6,3],[4,7,11]])
D = np.mat([[1,0,0],[0,1,0],[0,0,1]])
E = np.mat([[1,3],[5,7]])

mat_dict = {
    'A':A,
    'B':B,
    'C':C,
    'D':D,
    'E':E,
}

# Print shapes (dimensions row, col)
for key,mat in mat_dict.items():
    print(key, mat.shape)
```

```python
# Matrix multiplication requries that the number of columns in the first matrix (left-hand) be equivalent to number of 
# rows in the second (right-hand), such that m,n x n,k

from itertools import permutations

# Get all possible combinations with no replacement (no 'exponentials')
perms = permutations(list(mat_dict.keys()), 2)

# Check for valid combinations n1==n2
for combo in perms:
    lmat_shape = mat_dict[combo[0]].shape
    rmat_shape = mat_dict[combo[1]].shape
    lcolumns = lmat_shape[1]
    rrows = rmat_shape[0]
    if lcolumns == rrows:
        print('Valid Match Found:')
        print(combo)
    
```

## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```python
combos = [
    ('C', 'D'),
    ('A', 'E'),
    ('B', 'A'),
]

# find products
for combo in combos:
    out = np.matmul(mat_dict[combo[0]], mat_dict[combo[1]])
    print(combo)
    print(out)
    print(out.shape)
    print()
```

**If we take a look at C.shape (2,3) and D.shape (3,3); their product.shape is the number of rows in C and columns in D**
**This holds for A.shape (3,2) and E.shape(2,2); so we'd expect a 3x2 matrix and that's what we get!**

\begin{align}
G= 
\begin{bmatrix}
12 & 11 \\

```python
## Assuming that I is the identity array of matching size to G

G = np.mat([[12,11],[7,10]])
I = np.mat(np.identity(2))
print(I)
print(np.matmul(G,I))

# So, the identity matrix times itself is itself....er IG = G
# Can verify, [[12*1+12*0, 11*1+11*0],[7*0+7*1, 10*0+10*1]]
[[12*1+12*0, 11*1+11*0],[7*0+7*1, 10*0+10*1]]
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```python
H = np.mat([[12,11],[7,10]])
J = np.mat([[0,1,2],
           [7,10,4],
           [3,2,0]])

print(LA.det(H))
print(LA.det(J))
```

## 3.3 Find H^{-1} and then find J^{-1}

```python
print(LA.inv(H))
print()
print(LA.inv(J))
```

## 3.4 Find $HH^{-1}$ and then find $G^{-1}G$. Is $HH^{-1} == G^{-1}G$? Why or Why not?

```python
print('H x inv(H):')
print(np.matmul(H, LA.inv(H)))
print('inv(G) x G:')
print(np.matmul(LA.inv(G),G))
```

```python
np.matmul(H, LA.inv(H)) == np.matmul(LA.inv(G),G)
```

**Where an inverse matrix is defined, at matrix times its inverse it the identity matrix.  Because H and G are 2x2 matrices, their identity matrices are equivalent**
