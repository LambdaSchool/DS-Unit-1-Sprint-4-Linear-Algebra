## 1.1 Create a two-dimensional vector and plot it on a graph

```python
import matplotlib.pyplot as plt

two_d = [9, 2]

plt.arrow(0,0, two_d[0], two_d[1], head_width=.05, head_length=0.05, color ='red')
plt.xlim(-1,10)          
plt.ylim(-1,3)
plt.grid()
plt.show()
```

## 1.2 Create a three-dimensional vecor and plot it on a graph

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

three_d = [4, 7, 5]

vectors = np.array([[0, 0, 0, 4, 7, 5]])

X, Y, Z, U, V, W = zip(*vectors)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=1)
ax.set_xlim([0, 5])
ax.set_ylim([0, 8])
ax.set_zlim([0, 6])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

## 1.3 Scale the vectors you created in 1.1 by $5$, $\pi$, and $-e$ and plot all four vectors (original + 3 scaled vectors) on a graph. What do you notice about these vectors? 

```python
import math
import numpy as np

# Then multiply it by a scalar

two_d_five = np.multiply(5, two_d)

two_d_pi = np.multiply(math.pi, two_d)

two_d_e = np.multiply(-math.e, two_d)

# Plot the Scaled Vectors
plt.arrow(0,0, two_d[0], two_d[1], head_width=1, head_length=1, color ='red')
plt.arrow(0,0, two_d_five[0], two_d_five[1], head_width=1, head_length=1, color ='orange')
plt.arrow(0,0, two_d_pi[0], two_d_pi[1], head_width=1, head_length=1, color ='purple')
plt.arrow(0,0, two_d_e[0], two_d_e[1], head_width=1, head_length=1, color ='green')
plt.xlim(-28,50)
plt.ylim(-9,15)
plt.grid()
plt.show()
```

The scaled vectors are all collinear with the unscaled vector.

## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{e} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{f} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```python
g = [1, 1, 1, 8]
h = [3, 3, 3, 3]
norm_g = np.linalg.norm(g)
norm_h = np.linalg.norm(h)

print("norm_g:", norm_g)
print("norm_h:", norm_h)

if norm_g > norm_h:
    print("\nnorm_g is longer")
elif norm_h > norm_g:
    print("\nnorm_h is longer")
else:
    print("\nnorm_g and norm_h are the same length")
```

## 1.9 Show that the following vectors are orthogonal (perpendicular to each other):

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 0 \\ -1  \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 1 \\ \sqrt{2} \\ 1 \end{bmatrix}
\end{align}

```python
g = [1, 0, -1]
h = [1, 2**.5, 1]

if np.dot(g, h) == 0:
    print("g and h are perpendicular")
else:
    print("g and h are NOT perpendicular")
```

## 2.1 What are the dimensions of the following matrices? Which of the following can be multiplied together? See if you can find all of the different legal combinations.
\begin{align}
A = \begin{bmatrix}
1 & 2 \\
3 & 4 \\
5 & 6
\end{bmatrix}
\qquad
B = \begin{bmatrix}

\end{bmatrix}
\qquad
C = \begin{bmatrix}

\end{bmatrix}
\qquad
D = \begin{bmatrix}

\end{bmatrix}
\qquad
E = \begin{bmatrix}

\end{align}

```python
A_dim = [3, 2]
B_dim = [1, 3]
C_dim = [2, 3]
D_dim = [3, 3]
E_dim = [2, 2]

dims = [A_dim, B_dim, C_dim, D_dim, E_dim]
matrices = ["A", "B", "C", "D", "E"]

for i in range(len(dims)):
    for j in range(len(dims)):
        if dims[i][1] == dims[j][0]:
            print(matrices[i], "can be multipled by", matrices[j])
```

## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```python
A = np.array([[1, 2], [3, 4], [5, 6]])
B = np.array([2, 4, 6])
C = np.array([[9, 6, 3], [4, 7, 11]])
D = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]])
E = np.array([[1, 3], [5, 7]])

print(np.matmul(C, D))
print()
print(np.matmul(A, E))
print()
print(np.matmul(B, A))
```

The product of matrix multiplication has the same number of rows as the first matrix and the same number of columns as the second matrix.

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```python
H = np.array([[12, 11], [7, 10]])
J = np.array([[0, 1, 2], [7, 10, 4], [3, 2, 0]])

print(np.linalg.det(H))
print()
print(np.linalg.det(J))
```

## 3.3 Find H^{-1} and then find J^{-1}

```python
H_inv = np.linalg.inv(H)
J_inv = np.linalg.inv(J)

print(H_inv)
print()
print(J_inv)
```

## 3.4 Find $HH^{-1}$ and then find $G^{-1}G$. Is $HH^{-1} == G^{-1}G$? Why or Why not?

```python
# H = G, so I'll just use H for the variable name

print(np.matmul(H, H_inv))
print()
print(np.matmul(H_inv, H))
```

Square matrices have the same dimensions as their inverses. The product of a matrix and its inverse is just the identity matrix of the same dimension. Order does not matter when you are multiplying a matrix by its inverse; you get the same identity matrix either way.
