## 1.1 Create a two-dimensional vector and plot it on a graph

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
import matplotlib as mpl
from mpl_toolkits.mplot3d import axes3d
```

```python
purple = [2, 7]
plt.arrow(0, 0, purple[0], purple[1], head_width = .5, head_length = .5, color = 'purple')
axes = plt.gca();
axes.set_xlim(-8,8)
axes.set_ylim(-8,8)
axes.grid()
plt.title('2-D Vector')
```

## 1.2 Create a three-dimensional vecor and plot it on a graph

```python
red = [1, 2, 4]
blue = [2, 4, 6]
yellow = [1, 1, 2]

vectors = np.array([[0, 0, 0, 1, 2, 4],
                    [0, 0, 0, 2, 4, 6],
                    [0, 0, 0, 1, 1, 2]])

X, Y, Z, U, V, W = zip(*vectors) #zips all of the vectors into one. So I dont have to individually assign them variables.
fig = plt.figure()
ax = fig.add_subplot(111, projection= '3d') #creates 3d plot
ax.quiver(X, Y, Z, U, V, W, length=1)
ax.set_xlim([0, 8])
ax.set_ylim([0, 8])
ax.set_zlim([0, 8])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

## 1.3 Scale the vectors you created in 1.1 by $5$, $\pi$, and $-e$ and plot all four vectors (original + 3 scaled vectors) on a graph. What do you notice about these vectors? 

```python
scale_purple = [purple, np.multiply(5, purple), np.multiply(math.pi, purple), np.multiply((math.e*-1), purple)]
colors = ['purple', 'red', 'blue', 'green']
i=0
for arrow in scale_purple:
  plt.arrow(0,0, arrow[0], arrow[1],head_width=.5, head_length=0.5, color = colors[i])
  i+=1
  
axes = plt.gca();
axes.set_xlim(-10,10);
axes.set_ylim(-10,10);
axes.grid()

#The vectors are linearly dependent
```

## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

## 1.9 Show that the following vectors are orthogonal (perpendicular to each other):

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 0 \\ -1  \end{bmatrix}

\vec{h} = \begin{bmatrix} 1 \\ \sqrt{2} \\ 1 \end{bmatrix}

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```python
h = [[12,11], [7,10]]
j = [[0,1,2],[7,10,4],[3,2,0]]
print(np.linalg.det(h))
print(np.linalg.det(j))
```

## 3.3 Find H^{-1} and then find J^{-1}

```python
h_inv = np.linalg.inv(h)
j_inv = np.linalg.inv(j)
print(h_inv)
print(j_inv)
```

## 3.4 Find $HH^{-1}$ and then find $G^{-1}G$. Is $HH^{-1} == G^{-1}G$? Why or Why not?

```python
print(np.matmul(h, h_inv))
print(np.matmul(j_inv,j))
```
