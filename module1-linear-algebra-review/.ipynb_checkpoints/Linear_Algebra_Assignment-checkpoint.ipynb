# Part 2 - Matrices

## 2.1 What are the dimensions of the following matrices? Which of the following can be multiplied together? See if you can find all of the different legal combinations.
\begin{align}
A = \begin{bmatrix}
1 & 2 \\
3 & 4 \\
5 & 6
\end{bmatrix}
\qquad
B = \begin{bmatrix}
2 & 4 & 6 \\
\end{bmatrix}
\qquad
C = \begin{bmatrix}
9 & 6 & 3 \\
4 & 7 & 11
\end{bmatrix}
\qquad
D = \begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}
\qquad
E = \begin{bmatrix}
1 & 3 \\
5 & 7
\end{bmatrix}
\end{align}

```python
a = np.array([[1,2],[3,4],[5,6]])
b = np.array([[2,4,6]])
c = np.array([[9,6,3],[4,7,11]])
d = np.array([[1,0,0],[0,1,0],[0,0,1]])
e = np.array([[1,3],[5,7]])

dimensions = {'a': a.shape, 'b': b.shape, 'c': c.shape, 'd': d.shape, 'e': e.shape}
for key, value in dimensions.items():
    print(f'The shape of {key} is {value}')
    
for key, value in dimensions.items():
    for next_key, next_value in dimensions.items():
        if value[1] == next_value[0]:
            print(f'Matrix {key} can be multiplied with matrix {next_key}')
        else:
            continue
            
```

## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```python
cd = np.matmul(c,d)
ae = np.matmul(a,e)
ba = np.matmul(b,a)

print(f'CD is: \n', cd)
print(f'AE is: \n', ae)
print(f'BA is: \n', ba)

print(f'The shape of CD is {cd.shape}')
print(f'The shape of AE is {ae.shape}')
print(f'The shape of BA is {ba.shape}')
```

The dimension of the resulting matrix is the result of the number of rows of the first matrix and the number of columns of the second matrix.

# Part 3 - Square Matrices

## 3.1 Find $IG$ (be sure to show your work) 😃

\begin{align}
G= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\end{align}

```python
g = np.array([[12,11],[7,10]])
inverse = linalg.inv(g)
identity = np.matmul(inverse,g)
ig = np.matmul(identity,g)
ig
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```python
h = np.array([[12,11],[7,10]])
j = np.array([[0,1,2],[7,10,4],[3,2,0]])
print(f'The determinant of H is {linalg.det(h)}')
print(f'The determinant of J is {linalg.det(j)}')
```

## 3.3 Find H^{-1} and then find J^{-1}

```python
h_inverse = linalg.inv(h)
j_inverse = linalg.inv(j)
print(f'The inverse of H is: \n', h_inverse)
print(f'The inverse of J is: \n', j_inverse)
```

## 3.4 Find $HH^{-1}$ and then find $G^{-1}G$. Is $HH^{-1} == G^{-1}G$? Why or Why not?

```python
h_times_inverse = np.matmul(h,h_inverse).round(6)
g_times_inverse = np.matmul(inverse, g).round(6)

print(f'H multiplied by the inverse of H is: \n', h_times_inverse)
print(f'The inverse of G multiplied by G is: \n', g_times_inverse)
```

If a matrix has an inverse, as in the case of H and G, then the multplication between the matrix and its inverse is commutative and results in the identity matrix. In this case, it's the same since H and G have the same dimensions.
