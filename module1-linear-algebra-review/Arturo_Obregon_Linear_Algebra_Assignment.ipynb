## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```
g = np.sqrt((1**2)+(1**2)+(1**2)+(8**2))
h = np.sqrt((3**2)+(3**2)+(3**2)+(3**2))
print(g)
print(h)

if g > h:
  print('g is longer')
else:
  print('h is longer')
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
h = np.array([[12,11],
            [7,10]])

j = np.array([[0,1,2],
             [7,10,4],
             [3,2,0]])

hh = np.linalg.det(h)
jj = np.linalg.det(j)
print('Determinant for matrix H =',hh,'\n')
print('Determinant for matrix J =',jj,'\n')
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
hhh = np.linalg.inv(h)

jjj = np.linalg.inv(j)

print('The inverse of H is\n',hhh,'\n')
print('The inverse of J is\n',jjj)
```
