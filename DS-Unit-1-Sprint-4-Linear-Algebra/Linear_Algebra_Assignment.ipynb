## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```
import math

g = [1, 1, 1, 8]
h = [3, 3, 3, 3]

def norm(vector):
  return math.sqrt(sum(x**2 for x in vector))
  
print(norm(g), norm(h))

# g is longer. 8.18 > 6
```

## 3.1 Find $IG$ (be sure to show your work) 😃

\begin{align}
G= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}

```
# IG = [[1, 0], [0, 1]] dot [[12, 11], [7, 10]]
#    = [[1x12 + 0, 1x11 + 0], [1x7 + 0, 1x10 + 0]] = G

G = [[12, 11], [7, 10]]
I = np.identity(2)

np.dot(I, G)
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
H = G.copy()
J = [[0, 1, 2],[7, 10, 4], [3, 2, 0]]

print(np.linalg.det(H))
print(np.linalg.det(J))
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
print(np.linalg.inv(H))
print(np.linalg.inv(J))
```

## 3.4 Find $HH^{-1}$ and then find $J^{-1}J$. Is $HH^{-1} == J^{-1}J$? Why or Why not?

```
H_inv = np.linalg.inv(H)
J_inv = np.linalg.inv(J)

print(np.dot(H, H_inv))
print(np.dot(J, J_inv))

# Getting some floating point errors in here, first of all.
# But disregarding that, the two aren't equal, despite being identity
# matrices. That's because they have different dimensions.
```
