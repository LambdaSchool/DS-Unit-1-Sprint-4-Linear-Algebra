## 1.1 Create a two-dimensional vector and plot it on a graph

```
import numpy as np 
import matplotlib.pyplot as plt
v1 = np.array([1,2])
fig, ax = plt.subplots(figsize = (5,5))
plt.arrow(0, 0, v1[0], v1[1], head_width=.2, head_length=0.2, color = 'cyan')
plt.xlim(0,5)          
plt.ylim(0,5)
```

## 1.2 Create a three-dimensional vecor and plot it on a graph

```
from mpl_toolkits.mplot3d import Axes3D

vone = np.array([1,2,4])

vtwo = np.array([[0, 0, 0, 1, 2, 4]]) #need nested brackets to satisfy iterable requirement

X, Y, Z, U, V, W = zip(*vtwo)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=1)
ax.set_xlim([0, 5])
ax.set_ylim([0, 5])
ax.set_zlim([0, 5])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show();
```
