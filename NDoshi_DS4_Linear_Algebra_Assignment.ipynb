## 1.1 Create a two-dimensional vector and plot it on a graph

```
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```
# Create 2D vector
vec = [2,3]

# Plot the vector on a graph
plt.arrow(0,0,vec[0],vec[1],head_width=0.2,head_length=0.2,color='blue')
plt.xlim(0,5)
plt.ylim(0,5)
plt.title('Simple Vector')
plt.show()
```

## 1.2 Create a three-dimensional vecor and plot it on a graph

```
# Create 3D vector
vec_3d = [0.2,0.53,0.4]

# Plot 3D vector

# Import additional plotting library needed to plot 3D vector
from mpl_toolkits.mplot3d import Axes3D

vecs = np.array([[0,0,0,0.2,0.53,0.4]])

X,Y,Z,U,V,W = zip(*vecs)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=1)
ax.set_xlim([0, 0.75])
ax.set_ylim([0, 0.75])
ax.set_zlim([0, 0.75])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

## 1.3 Scale the vectors you created in 1.1 by $5$, $\pi$, and $-e$ and plot all four vectors (original + 3 scaled vectors) on a graph. What do you notice about these vectors? 

```
# Import Math library to use below
import math

# Take vec from above and multiple by scalers listed above
vec1 = np.multiply(5,vec)
vec2 = np.multiply(math.pi,vec)
vec3 = np.multiply(math.e, vec)

# Plot the scaled vectors on a graph
plt.arrow(0,0,vec1[0],vec1[1],head_width=0.2,head_length=0.2,color='red')
plt.arrow(0,0,vec2[0],vec2[1],head_width=0.2,head_length=0.2,color='green')
plt.arrow(0,0,vec3[0],vec3[1],head_width=0.2,head_length=0.2,color='yellow')
plt.arrow(0,0,vec[0],vec[1],head_width=0.2,head_length=0.2,color='blue')
plt.xlim(0,20)
plt.ylim(0,20)
plt.title('Scaled Vectors')
plt.show()
```
