## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
H = [[12,11],
    [7,10]]
H_determinant = np.linalg.det(H)
H_determinant

print('H_determinant: \n', H_determinant)

J = [[0,1,2],
    [7,10,4],
    [3,2,0]]
J_determinant = np.linalg.det(J)
print('J determinant: \n', J_determinant.round())
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
inverse_of_H = np.linalg.inv(H)
inverse_of_J = np.linalg.inv(J)
print('Inverse of H: \n', inverse_of_H)
print('Inverse of J: \n', inverse_of_J)
```

## 3.4 Find $HH^{-1}$ and then find $J^{-1}J$. Is $HH^{-1} == J^{-1}J$? Why or Why not?

```
HH_inverse = np.matmul(H,inverse_of_H)
JJ_inverse = np.matmul(J,inverse_of_J)
print("H times inverse of H: \n", HH_inverse.round())
print("J times inverse of J: \n", JJ_inverse.round())
#The number of non-zeros is the rank of the matrix. HH_inverse has a rank of 2.
#JJ_inverse has a rank of 3. Rank is the dimension of the vector space generated.
```
