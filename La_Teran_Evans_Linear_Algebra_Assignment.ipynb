## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
# find the determinants
h = np.array([[12,11], [7,10]])
j = np.array([[0,1,2], [7,10,4], [3,2,0]])

print(np.linalg.det(h))
print(np.linalg.det(j))
```

```
# find the inverse? 

h = np.array([[12,11], [7,10]])
j = np.array([[0,1,2], [7,10,4], [3,2,0]])

print(h)
print(np.linalg.inv(h))
print(" ")
print(j)
print(np.linalg.inv(j))
```
