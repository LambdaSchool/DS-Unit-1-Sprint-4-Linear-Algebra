```python
rating_final['average'] = (rating_final['rating']+rating_final['food_rating']+rating_final['service_rating'])/3
```

```python
rating_final.head()
```

```python
rating_final_average = rating_final.drop(columns=['rating','food_rating','service_rating'])
rating_final_average.head()
```

```python
rating_final_average_personality = rating_final_average.merge(user_profile[['userID', 'personality']])
```

```python
rating_final_average_personality = rating_final_average_personality.merge(payment_methods[['placeID', 'Rpayment']])
```

```python
rating_final_average_personality.head()
```

```python
crosstab_average_personality = pd.crosstab(rating_final_average_personality['personality'],
                                           rating_final_average_personality['average'] )
crosstab_average_personality.T.head()
```

```python
crosstab_average_personality.T.plot(kind='line')
```

```python
crosstab_average_paymentmethod = pd.crosstab(rating_final_average_personality['Rpayment'],
                                           rating_final_average_personality['average'] )
crosstab_average_paymentmethod.T.head()
```

```python
crosstab_average_paymentmethod.T.plot(kind='line')
```

```python
pivot_average_personality = pd.pivot_table(rating_final_average_personality, values='average', index=['personality', 'userID'])
pivot_average_personality.head(20)
```
