```python
def CalculateCovariance(x):
    meanX = np.mean(x, axis=0)
    lenX = x.shape[0]
    x = x - meanX
    covariance = x.T.dot(x)/lenX
    return covariance
```

```python
CalculateCovariance(real_estate)
```
